```python
import os, sys
from scipy.misc import imread,imresize,imsave
```

```python
def resize(path1,path2):
    for item in os.listdir(path1):
        img = imread(path1+item)
        shape1 = img.shape[0]
        shape2 = img.shape[1]
        diff = abs(shape1-shape2)//2
        print(img.shape)
        if(shape2>shape1):
            newimg = img[:,diff:shape2-diff]
        else:
            newimg = img[diff:shape1-diff,:]
        newShape1 = int(0.25*newimg.shape[0])
        newShape2 = int(0.25*newimg.shape[1])
        newimg = newimg[newShape1:newimg.shape[0]-newShape1,newShape2:newimg.shape[1]-newShape2]
        print(newimg.shape)
        newimg = imresize(newimg,size=[224,224],interp = 'bicubic',mode=None)
        imsave(path2+item,newimg)
```

```python
path1 = 'data/DIV2K_train_HR/'
path2 = 'data/DIV2K_train_HR_resized_crops/'
resize(path1,path2)
```

```python
path1 = 'data/DIV2K_valid_HR/'
path2 = 'data/DIV2K_valid_HR_resized_crops/'
resize(path1,path2)
```

```python
def resize_BSD(path1,path2):
    for item in os.listdir(path1):
        img = imread(path1+item)
        shape1 = img.shape[0]
        shape2 = img.shape[1]
        diff = abs(shape1-shape2)//2
        print(img.shape)
        if(shape2>shape1):
            newimg = img[:,diff:shape2-diff]
        else:
            newimg = img[diff:shape1-diff,:]
        print(newimg.shape)
        diff1 = newimg.shape[0] - 224
        diff2 = newimg.shape[1] - 224
        newimg = newimg[diff1:newimg.shape[0],diff2:newimg.shape[1]]
        print(newimg.shape)
        imsave(path2+item,newimg)
```

```python
path1 = 'data/BSDS300/images/'
path2 = 'data/BSDS300_cropped/'
resize_BSD(path1,path2)
```
